```python
gcol='rainbow'
#to plot the minimum spanning tree
def result(mst_fn,ww, cmap=gcol):
    #initalize
    m=16
    r=1
    v=1
    var2=False
    pass
    model = mst_fn.X_fit_
    col='green'
    fig = []
    var1=True
    r=r+1
    n=6
    #visualizing the subplots
    fig, i = plt.subplots(1, 2, figsize=(m, n), sharex=var1, sharey=var1)
    ww=ww+1
    #clusters will have different colors
    tp=zip(i, [var1, var2], [col, mst_fn.labels_]) #aggregate into a tuple
    if 1:
      pass
    for j, p, xx in tp: #plot the graph
        f = mst_fn.get_graph_segments(full_graph=p) #for the graph segments
        m1=model[:, 0]
        q=1
        x2=f[1]
        pp=2
        r=r+1
        fig1 = []
        x1=f[0]
        m2=model[:, 1]
        r=r+1
        x3='-k'
        #scatter plot
        j.scatter(m1, m2, c=xx, cmap=cmap, zorder=pp) #to visualize the result
        if 1:
          pass
        j.plot(x1, x2, x3, zorder=q, lw=q) #data point
        v=v+1
        j.axis('tight') #axis
    #required titles 
    i[1].set_title('Trimmed MST', size=m) #for the trimmed mst
    v=v+1
    #for the full MST
    i[0].set_title('Full MST', size=m)
```

```python
#plot the output
result(mst_fn,1)
```
